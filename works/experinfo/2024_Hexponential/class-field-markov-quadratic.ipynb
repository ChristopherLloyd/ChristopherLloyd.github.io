```python
### ### LIBRARIES ### ###

import numpy as np
from cypari import pari
# import matplotlib.pyplot as plt
```

## Finding Markov Quadratics

```python
m = 5 #1,2,5,13,29
D = (9*m**2-4) # discriminant
funD = pari.coredisc(D) # discriminant of Q(sqrt(D))
print(D, funD)

wD = pari.quadgen(D) # the omega_D
wfunD = pari.quadgen(funD) # the omega_funD
print(wD*1., wfunD*1.)

print('isfundamental(D)=', pari.isfundamental(D)==1)
print('trace(wD) = ', pari.trace(wD), '\t', 'norm(wD) =', pari.norm(wD), '\t',\
      'charpoly(wD)=', pari.charpoly(wD))
print('trace(wfunD) = ', pari.trace(wfunD), '\t', 'norm(wfunD) =', pari.norm(wfunD), '\t',\
      'charpoly(wfunD)=', pari.charpoly(wfunD))

print('sqrt(D)=', pari.contfrac(pari.sqrt(D)), '\n', 'wD =', pari.contfrac(wD*1.))
print('sqrt(funD)=', pari.contfrac(pari.sqrt(funD)), '\n', 'wfunD =', pari.contfrac(wfunD*1.))

pari.quadclassunit(funD)
```

```python

bnf = pari.bnfinit(pari.charpoly(wfunD))
pari.bnrclassfield(bnf)
```
