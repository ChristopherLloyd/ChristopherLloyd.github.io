## Recuperating $\eta^4(6\tau)$ from LMFBD 

at https://www.lmfdb.org/ModularForm/GL2/Q/holomorphic/36/2/a/a/

We let $q=\exp(i 2\pi n \times \tau)

We define $\psi(n)$ by $\eta^4(6\tau) = q\prod_1^\infty (1-q^{6n})^4 = \sum_1^\infty \psi(n)q^{n/4}$.

We define $hexp(\tau) = \int \eta^4(\tau) d\tau = C \times \sum_1^\infty \frac{1}{n} \psi(n) \exp(i 2\pi n \times \tau)$

```python
# https://www.lmfdb.org/ModularForm/GL2/Q/holomorphic/36/2/a/a/

[Level, weight, char] = [36,2,pari.Mod(1,36)]
print(42*char, '\n')

mf = pari.mfinit([Level, weight, char],0)
print(mf, '\n')

lf = pari.mfeigenbasis(mf)
print(lf, '\n')

eta_4_6 = lf[0]
print(eta_4_6, '\n')
```

```python
def eta_4_6_coef(n):
    return pari.mfcoefs(eta_4_6, n)
print(eta_4_6_coef(43))

def eta_4_6_ser(n):
    return pari.Ser(pari.mfcoefs(eta_4_6, n),pari('q'))
print(eta_4_6_ser(43))

def hexp_ser(n):
    eta_4_6_ser_n = eta_4_6_ser(n)
    return pari.intformal(eta_4_6_ser_n/pari('q'))
print(hexp_ser(43))
```

### partial sums of the hexponential cusp compactification

$\partial hexp(0) = \sum_1^\infty \frac{1}{n} \psi(n)$ 
and of 
$\partial hexp(a/b) = \sum_1^\infty \frac{1}{n} \psi(n) \exp(i 2\pi n \times a/b)$

```python
def partial_sum_series(degree):
    coeffs = pari.mfcoefs(eta_4_6, degree)
    partialsum = 0 #coeffs[1]
    for k in range(1, degree, 6):
        partialsum += coeffs[k]/k #6*coeffs[k]/(k-1)
        print(k, coeffs[k]/k, partialsum*1.)
    return(partialsum)

partial_sum_series(200)
```

## Padé approximants for $\eta^4$ and its primitive $hexp = \int \eta^4(z)dz$

```python
# pari.allocatemem(4)
N = 10

print(eta_4_6_ser(N))
print(hexp_ser(N))

# print(pari.contfracinit(eta_4_6_ser(N)) )

print(pari.bestapprPade( eta_4_6_ser(N) ) )
print(pari.bestapprPade( hexp_ser(N) ) )

for k in range(N):
    print ( (pari.bestapprPade( eta_4_6_ser(k)).subst('q', 1) ) )

print('\n')

for k in range(N):
    print ( (pari.bestapprPade( hexp_ser(k)).subst('q', 1) )*1. )
```
